```python
import json
import numpy as np

# file_paths = ['/content/train_spider.json', '/content/train_others.json']

from google.colab import drive
drive.mount('/content/drive')
spiderFilePath = '/content/drive/MyDrive/spider/train_spider.json'
othersFilePath = '/content/drive/MyDrive/spider/train_others.json'

#initialize list to store sentences from each file
trainSpider = []
trainOthers = []

# Read data from train_spider.json
with open(spiderFilePath, 'r') as f_spider:
  trainSpider = json.load(f_spider)

print(len(trainSpider))

# Read data from train_spider.json
with open(othersFilePath, 'r') as f_others:
  trainOthers = json.load(f_others)

print(len(trainOthers))


# combine train_spider and train_others
allData = np.array(trainSpider + trainOthers)
```

```python
# Now extract natual language queries (NLQs), i.e. questions, from the dataset from the dataset

allNLQs = np.array( [data['question'] for data in allData] )
allNLQsTokens = np.array( [data['question_toks'] for data in allData], dtype=object )
allQueryTokens = np.array( [data['query_toks'] for data in allData], dtype=object )
```

```python
import stanza
stanza.download('en')
nlpStanza = stanza.Pipeline('en')

# posTagsOfNLQs = []
# for question in allNLQs:
#   # Process tasks
#   doc = nlpStanza(question)
#   # Perform NLP tasks
#   questionTokensPos = ''
#   for sentence in doc.sentences:
#     for word in sentence.words:
#       questionTokensPos += word.text + '/' + word.xpos + '/ '

#   posTagsOfNLQs.append(questionTokensPos)


# Extracting pos tags of questins
def extractPosTagsOfQuestions(listOfNLQs):
  posTagsOfNLQs = []
  cnt = 1
  for question in listOfNLQs:
    print('item no: ', cnt)
    cnt += 1
    # process tasks
    doc = nlpStanza(question)
    # Perfomr NLP tasks
    questionTokensAndPos = ''
    for sentence in doc.sentences:
      for word in sentence.words:
        questionTokensAndPos += word.text + '/' + word.xpos + '/ '

    posTagsOfNLQs.append(questionTokensAndPos)

  return posTagsOfNLQs


# Create file and write constructed data
def writePosTagsToFile(fileAndPath, aList):
  with open(fileAndPath, 'w') as fileHandle:
    for sentenceWithPos in aList:
      fileHandle.write(sentenceWithPos)
      fileHandle.write('\n')


posTagsOfNLQs = extractPosTagsOfQuestions(allNLQs)

filePathAndfile = '/content/drive/MyDrive/CS559_shared/implementation/question_pos_tags.txt'
writePosTagsToFile(filePathAndfile, posTagsOfNLQs)

print('extracting POS of NLQs are completed.')
```
